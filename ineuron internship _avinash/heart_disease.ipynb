## Heart Disease Diagnostic Analysis

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb
```

```python
data=pd.read_csv("heart_disease_dataset.csv")
```

```python
data.head()
```

```python
data.info()
```

There are thirteen features in Dataset
age: The person's age in years

sex: The person's sex (1 = male, 0 = female)

cp: The chest pain experienced (Value 1: typical angina, Value 2: atypical angina, Value 3: non-anginal pain, Value 4: asymptomatic)

trestbps: The person's resting blood pressure (mm Hg on admission to the hospital)

chol: The person's cholesterol measurement in mg/dl

fbs: The person's fasting blood sugar (> 120 mg/dl, 1 = true; 0 = false)

restecg: Resting electrocardiographic measurement (0 = normal, 1 = having ST-T wave abnormality, 2 = showing probable or definite left ventricular hypertrophy by Estes' criteria)

thalach: The person's maximum heart rate achieved

exang: Exercise induced angina (1 = yes; 0 = no)

oldpeak: ST depression induced by exercise relative to rest

slope: the slope of the peak exercise ST segment (Value 1: upsloping, Value 2: flat, Value 3: downsloping)

ca: The number of major vessels (0-3)

thal: A blood disorder called thalassemia (3 = normal; 6 = fixed defect; 7 = reversable defect)

num: Heart disease (0 = no, 1 = yes)

## Percentage of people having Heart Disease

```python
data['num'].value_counts()
```

```python
#Converting Numerical Data into Categorical Data
def heart_disease(x):
    if x==0:
        return 'Absense'
    if x==1:
        return "Present"
```

```python
data['disease']=data['num'].apply(heart_disease)
data
```

```python
a=data['disease'].value_counts()
```

```python
sb.countplot(data['disease'])
```

```python
plt.figure(figsize=(10,7))
plt.pie(a, labels=['Absence','Presence'], autopct='%0.0f%%')
plt.title('Heart Disease Population %', fontsize=20)
plt.show()
```

46% of the population are suffering from heart disease and it is major concern.

```python
plt.figure(figsize=(8,6))
plt.xticks(rotation=0, fontsize=18)
sb.countplot(x=data['sex'],hue='num',data=data)
plt.xlabel("0: FEMALE, 1: MALE",fontsize=18)
plt.ylabel("count",fontsize=18)
plt.title('Gender v/s target\n')
plt.show()
```

Males have higher chances of having heart disease than females.

```python
#Converting Numerical Data into Categorical Data

def gender(row):
    if row==1:
        return 'Male'
    elif row==0:
        return 'Female'
```

```python
data['sex']=data['sex'].apply(gender)
data.head()
```

```python
#Converting Numerical Data into Categorical Data

def age_range(row):
    if row>=29 and row<40:
        return 'Young Age'
    elif row>=40 and row<55:
        return 'Middle Age'
    elif row>55:
        return 'Elder Age'
```

```python
data['Age_Range']=data['age'].apply(age_range)
data.head()
```

```python
plt.figure(figsize=(10,7))
sb.swarmplot(x='Age_Range', y='age', hue='sex', data=data, order=['Young Age','Middle Age','Elder Age'], palette='Oranges_r')
plt.title('Gender Based Age Category', fontsize=17)
plt.xlabel('Age Category', fontsize=15)
plt.ylabel('Age', fontsize=15)
plt.show()
```

```python
plt.figure(figsize=(7,5))
sb.countplot(x=data['disease'], hue='sex', data=data, palette='BuGn_r')
plt.xlabel('Heart_Disease', fontsize=15)
plt.ylabel('Count',fontsize=15)
plt.legend(labels=['Male','Female'])
plt.title('Heart Disease Based on Gender',fontsize=17)
plt.show()
```
